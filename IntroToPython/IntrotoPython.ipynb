## How to manipulate columns with Pandas

```python
# Let's first display our top 5 rows for reference 
df.head()
```

### Insert/Add Columns

```python
# Now lets make a column called Total Score that is based on the Sum of the Math, Reading, and Writing scores
df['Total'] = df['math score'] + df['reading score'] + df['writing score']
df.head()
```

```python
#If we wanted to insert the column in a specific index in the DataFrame we would use
#In Python, the indexing begins at 0, so whenever you begin counting, remember the first item is number 0
df.insert(0, 'total', df['math score'] + df['reading score'] + df['writing score'])
df.head()
```

### Rename Columns

```python
# To rename a columns we will use
df.rename(columns={'Total':'Sum Of Scores'},inplace = True)
df
```

### Drop/Delete Columns

```python
# To drop unwanted columns we will use
df.drop('total',axis = 1,inplace = True)
df
```

```python
# We can also drop columns by using their indexed location, so here we will drop the race/ethnicity column and the Sum column
df.drop(df.columns[[1,8]],axis = 1, inplace = True)
df
```

# Plotting with Seaborn

## Scatterplots!

```python
# First let's make a scatterplot to see if there is any relationship between a student's math score and reading score
sb.scatterplot(data=df,x='math score',y='reading score')
```

```python
# Now we will see what the scatterplot looks like when the dots are colored by the type of education the students' parents have
sb.scatterplot(data=df,x='math score',y='reading score',hue='parental level of education')
```

```python
# Now let's change the color of the dots, based on what type of lunch the student has
sb.scatterplot(data=df,x='math score',y='reading score',hue='lunch')
```

## Histograms!

```python
#Now let's make some histograms to see the distribution of the students' exam scores
#Math Distribution
sb.histplot(data=df,x='math score')
```

```python
#Reading Distribution
sb.histplot(data=df,x='reading score')
```

```python
#Writing Distribution
sb.histplot(data=df,x='writing score')
```

```python
#We can flip a histogram to the y-axis as well
#Writing Distribution
sb.histplot(data=df,y='writing score')
```

```python
#We can also change the number of bins, the bins are the rectangles and finding a sweet spot can help with visualization
#Low # of bins
sb.histplot(data=df,x='writing score',bins=3)
```

```python
#High number of bins
sb.histplot(data=df,x='writing score',bins=150)
```

```python
#Sweet Spot # of bins
sb.histplot(data=df,x='writing score',bins=30)
```

```python
#We can also add a Kernel Density Estimate to these plots to aid in understanding the shape of the histogram
sb.histplot(data=df,x='writing score',bins=30,kde=True)
```

```python
#We can plot all of the scores at the same time, but this often gets noisy and a bit hard to understand
sb.histplot(data=df)
```

```python
#To help with this, we can stack the data to better see the data separation
sb.histplot(data=df,multiple='stack')
```

```python
#We can also change the shape of the histogram by using the element poly, which will more accurately model changes
sb.histplot(data=df,element='poly')
```

```python
#We can again view the distribution of scores based on lunch and parental education level
#Lunch hue
sb.histplot(data=df, x='math score', hue='lunch')
```

```python
#We can again view the distribution of scores based on lunch and parental education level
#Parental Education Level hue
sb.histplot(data=df, x='math score', hue='parental level of education')
```

```python
#Oncd again, due to the number of different levels, the data gets a bit noisy, so we can use the "stack" parameter
#to see the separation in the data a bit better
#Parental Education Level hue
sb.histplot(data=df, x='math score', hue='parental level of education',multiple='stack')
```

## Barplots!

```python
#We can make barplots to see the counts within categorical variables (such as parental level of education)
#The barplots will model the average score for each of the different catergories
sb.barplot(data=df,x='parental level of education',y='math score')
```

```python
#We can once again base this same plot based on the separation in genders, to see which has a higher average grade
sb.barplot(data=df,x='parental level of education',y='math score',hue='gender')
```

## Heatmaps!

```python
#An important tool to use in data analytics is the heatmap, in which it is very easy to model correlation between variables
#The first step to this is to use the pandas .corr function to get correlation values from your dataframe
correlation_df = df.corr(min_periods=3)
correlation_df
```

```python
#Next we can plot this data into a heatmap to see the correlation between the different scores
#Here we see that all of the scores are strongly correlated, but reading and writing scores have the highest correlation
#Meaning if a student scores high on the reading exam they will most likely score high on the writing exam (and vice-versa) 
sb.heatmap(correlation_df, annot=True, linewidths=0.5,cmap='YlGnBu')
```

## Pairplots

```python
#Lastly we will look into pairplots, which create a matrix to identify correlations through pairing scatterplots & histograms
sb.pairplot(df)
```

```python
#Once again we can separate these by categorical values (such as gender/lunch)
#Lunch values
sb.pairplot(df, hue='lunch')
```

```python
#Gender values
sb.pairplot(df,hue='gender')
```

```python
#Test Prep Course
sb.pairplot(df,hue='test preparation course')
```

```python
#To make the matrix cleaner we can also cut the mirrored half off
sb.pairplot(df,hue='test preparation course',corner=True)
```

```python
#We can also control how the diagonal is plotted, above we see that it is defaulting to a kde plot
#Now we will replot it as a histogram
sb.pairplot(df,hue='test preparation course',diag_kind='hist')
```
